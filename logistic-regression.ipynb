```python
### STEP 1 : IMPORT REQUIRED LIBRARIES ###
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
### STEP 2 : READ_CSV FILE
```

```python
df = pd.read_csv('./social_ads.csv')
```

```python
df
```

```python
### STEP 3 : CHECK ANY NULL VALUES
```

```python
print(df.isnull().sum()) 
```

```python
### STEP 4 : REMOVE NULL VALUES 
```

```python
df = df.dropna()
```

```python
df
```

```python
### STEP 5 : INITIALIZE X AND Y 
### X -> INDEPENDENT
### Y -> DEPENDENT
```

```python
X = df[['Age']]

y = df[['Purchased']]
```

```python
X
```

```python
y
```

```python
### STEP 6 : SPLITING DATA INTO TEST AND TRAINING
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=12)
```

```python
X_train , y_train
```

```python
X_test , y_test
```

```python
### SIZE OF DATA SPLITTED

print("X_test : " ,X_test.shape)
print("y_test : " ,y_test.shape)
print("X_train : " ,X_train.shape)
print("y_train : " ,y_train.shape)
```

```python
### STEP 7 : INITIALIZING MODEL
```

```python
model = LogisticRegression()
```

```python
model
```

```python
### STEP 8 : TRAIN THE MODEL
```

```python
model.fit(X_train, y_train)
```

```python
### STEP 9 : MAKE PREDICTION
```

```python
y_pred =  model.predict(X_test)
```

```python
y_pred
```

```python
plt.scatter(X_test, y_test, color='blue')  # Actual data
plt.plot(X_test, y_pred, color='red')  # Predicted data
plt.xlabel('Age')
plt.ylabel('Purchased')
plt.legend()
plt.show()
```

```python
### STEP 10 : PLOT THE SIGMOID CURVE FOR LOGISTIC REGRESSION ###

# Import necessary libraries (already imported above)
# import matplotlib.pyplot as plt
# import numpy as np

# Create a range of Age values for smooth curve
X_range = np.linspace(X['Age'].min(), X['Age'].max(), 300).reshape(-1, 1)

# Get predicted probabilities for the sigmoid curve
y_prob = model.predict_proba(X_range)[:, 1]  # Probability of Purchased=1

# Plot the sigmoid curve
plt.figure(figsize=(10, 6))
plt.plot(X_range, y_prob, color='red', linewidth=2, label='Sigmoid Curve (Probability)')

# Plot the actual data points
plt.scatter(X_test, y_test, color='blue', label='Actual Data', alpha=0.7)

# Add decision boundary (where probability = 0.5)
# Find the age where probability crosses 0.5
threshold_age = X_range[np.argmin(np.abs(y_prob - 0.5))][0]
plt.axvline(x=threshold_age, color='green', linestyle='--', label=f'Decision Boundary (Age ≈ {threshold_age:.1f})')

# Labels and title
plt.xlabel('Age')
plt.ylabel('Probability of Purchase')
plt.title('Logistic Regression Sigmoid Curve')
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```
