```python
import matplotlib.pyplot as plt

labels = ["IBM", "SAP", "Google", "Microsoft", "X"]
mcap = [150, 125, 590, 450, 100]

# explode length must match mcap length
explode = [0, 0, 0.3, 0, 0]

plt.pie(mcap, labels=labels, explode=explode, autopct='%1.1f%%')
plt.title("Market Cap Distribution")
plt.show()
```

```python
import numpy as np

india = [2.04, 2.10, 2.29, 2.65, 2.70, 2.87, 2.67, 3.17, 3.39, 3.73]
china = [10.5, 11.0, 11.2, 12.0, 13.6, 14.3, 14.7, 17.7, 18.0, 18.5]
us = [17.5, 18.0, 18.6, 19.5, 20.5, 21.4, 20.9, 23.0, 25.0, 26.5]

years = [2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022, 2023]

plt.title("GDP Comparison (20142023)")
plt.bar(years,india,color="#fcb490",label="India")
plt.bar(years,china,color="#fa0707",label="China")
plt.bar(years,us,color="#0c1345",label="US")

plt.legend()
plt.show()
```

```python
# Import matplotlib for plotting graphs
import matplotlib.pyplot as plt

# Import numpy for handling numeric arrays and positioning bars
import numpy as np


# -------------------------------
# 1. DATA PREPARATION
# -------------------------------

# List of years (X-axis values)
years = [2014,2015,2016,2017,2018,2019,2020,2021,2022,2023]

# GDP values for each country (in Trillion USD)
# These lists must match the years list one-to-one
india = [2.04,2.10,2.29,2.65,2.70,2.87,2.67,3.17,3.39,3.73]
china = [10.5,11.0,11.2,12.0,13.6,14.3,14.7,17.7,18.0,18.5]
usa   = [17.5,18.0,18.6,19.5,20.5,21.4,20.9,23.0,25.0,26.5]


# -------------------------------
# 2. CALCULATING X-POSITIONS
# -------------------------------

# Create array of positions for each year (0,1,2,3...)
x = np.arange(len(years))
print(x)

# Width of each bar
# Since we have 3 bars per year, we need to keep them narrow
width = 0.25


# -------------------------------
# 3. CREATING THE BAR GRAPH
# -------------------------------

# Set the size of the figure (width, height)
# plt.figure(figsize=(19,6))

# Plot each country's GDP as bars
# We shift each dataset slightly left/right so bars dont overlap
plt.bar(x - width, india, width, label="India")
plt.bar(x, china, width, label="China")
plt.bar(x + width, usa, width, label="USA")


# -------------------------------
# 4. CUSTOMIZING AXES AND LABELS
# -------------------------------

# Set year labels on X-axis
plt.xticks(x, years)

# X-axis label
# plt.xlabel("Year")

# Y-axis label
# plt.ylabel("GDP (Trillion USD)")

# Main title of the graph
plt.title("GDP Comparison of India, China, and USA (20142023)")


# -------------------------------
# 5. ADDITIONAL IMPROVEMENTS
# -------------------------------

# Add legend to identify which bar belongs to which country
# plt.legend()

# Add horizontal grid lines (helps readability)
# plt.grid(axis="y")


# -------------------------------
# 6. DISPLAYING THE GRAPH
# -------------------------------

# Show the final bar chart on the screen
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

years = [2014,2015,2016,2017,2018,2019,2020,2021,2022,2023]

india = [2.04,2.10,2.29,2.65,2.70,2.87,2.67,3.17,3.39,3.73]
china = [10.5,11.0,11.2,12.0,13.6,14.3,14.7,17.7,18.0,18.5]
usa   = [17.5,18.0,18.6,19.5,20.5,21.4,20.9,23.0,25.0,26.5]

x = np.arange(len(years))
print("Arranged length of years : ",x)
print("Length of Years is :  ", len(years))


width = 0.25

plt.bar(x - width, india, width, label="India", color="orange")
plt.bar(x, china, width, label="China", color="red")
plt.bar(x + width, usa, width, label="USA" , color="blue")


plt.show()
```
