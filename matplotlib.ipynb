```python
import matplotlib.pyplot as plt

labels = ["IBM", "SAP", "Google", "Microsoft", "X"]
mcap = [150, 125, 590, 450, 100]

# explode length must match mcap length
explode = [0, 0, 0.3, 0, 0]

plt.pie(mcap, labels=labels, explode=explode, autopct='%1.1f%%')
plt.title("Market Cap Distribution")
plt.show()
```
